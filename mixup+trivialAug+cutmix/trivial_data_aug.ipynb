```python
import os 
import numpy as np 
from torchvision import datasets, transforms
import torch
import matplotlib.pyplot as plt
from collections import Counter

data_dir = './imbalanced/imbalanced'
train_dir = os.path.join(data_dir,'train/')

data_transform = transforms.Compose([transforms.Resize([224,224]), transforms.ToTensor(),transforms.ConvertImageDtype(torch.uint8)])
train_data = datasets.ImageFolder(train_dir, transform = data_transform)

# get number of sample in each subset
num_nv = dict(Counter(train_data.targets))[0]
num_v = dict(Counter(train_data.targets))[1]

nv_dset = torch.utils.data.Subset(train_data, range(0, num_nv))
dataloader = torch.utils.data.DataLoader(nv_dset, batch_size=num_nv)
images, labels = next(iter(dataloader))

# path to save augmented pitures
path = os.path.join(data_dir, 'train_augmented/Non Venomous')
if (os.path.isdir(path)): None
else: os.mkdir(path) 

# set number of augmentation for each sample
num_aug = 3

augmenter = transforms.TrivialAugmentWide()
for i in range(len(images)):
    img_ori = images[i]
    
    for j in range(num_aug):
        imgs = augmenter(img_ori)
        imgs = np.transpose(imgs.numpy(), (1, 2, 0))
        plt.imsave(path + '/img' + str(i) + '_' + str(j) + '.jpg',imgs)
        
    img_ori = np.transpose(img_ori.numpy(), (1, 2, 0))
    plt.imsave(path +'/img' + str(i) + '.jpg',img_ori) 
    



```
